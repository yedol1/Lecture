```python
import openpyxl

def write_excel_tmp(filename,sheetname,listdata): 
    excel_file=openpyxl.Workbook() # 엑셀 실행
    excel_sheet=excel_file.active # 파일 시트 활성화
    excel_sheet.column_dimensions['A'].width = 100 # 엑셀 행 이름 & 넓이
    excel_sheet.column_dimensions['B'].width = 20
    
    if sheetname != '':
        excel_sheet.title=sheetname # 만약 시트이름이 default값이 아니면 시트의 title을 시트이름으로 변경
    for item in listdata:
        excel_sheet.append(item) # listdata에 있는 항목들을 엑셀 시트에 추가
    excel_file.save('tmp.xlsx') # 저장
    excel_file.close() # 닫기
```

```python
import requests
from bs4 import BeautifulSoup

product_list=list()

for page_num in range(10):
    if page_num==0:
        res=requests.get('https://davelee-fun.github.io/')
    else:
        res=requests.get('https://davelee-fun.github.io/',str(page_num+1))
    soup=BeautifulSoup(res.content, 'html.parser')
    
    data=soup.select('div.card')
    for item in data:
        product_name=item.select_one('div.card-body h4.card-text')
        product_date=item.select_one('div.wrapfooter span.author-meta span.post-date')
        product_info=[product_name.get_text().strip(),product_date.get_text()]
        product_list.append(product_info)
write_excel_tmp('tmp.xlsx','상품정보',product_list)
```

```python
import requests
client_id='X_JRTQsqCYhl9NncUKCg'
client_pw='OSqB4Sfu_F'

start_num=1
num=0

excel_file=openpyxl.Workbook() # 엑셀 실행
excel_sheet=excel_file.active # 파일 시트 활성화
excel_sheet.column_dimensions['A'].width = 3 # 엑셀 행 이름 & 넓이
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['rank','title','link'])
for index in range(10): # display 최대가 100개인것을 for문을 통하여 원하는 갯수만큼으로 즐길수있다
    start_number=start_num+(index*100) # 1~100 101~200 ...
    naver_open_api='https://openapi.naver.com/v1/search/shop.json?query=아이패드&display=100&start='+str(start_number) # 1 , 101 , 201 , ... 
    header_params={'X-Naver-Client-Id':client_id,'X-Naver-Client-Secret':client_pw}
    res=requests.get(naver_open_api, headers=header_params)
    
    if res.status_code==200:
        data=res.json()
        #print(data['items'])
        for item in data['items']:
            num+=1 # num 1부터 100까지 올라가고 for문 한번돌면 리셋 , 101부터 200까지 올라가고 for문 한번돌면 리셋 , ...
            excel_sheet.append([num ,item['title'], item['link']])
        
    else :
        print("error:",res.status_code)
excel_file.save('shopping_with_아이패드.xlsx')
excel_file.close()
```
