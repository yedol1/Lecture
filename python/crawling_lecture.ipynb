```python
import openpyxl

def write_excel_tmp(filename,sheetname,listdata): 
    excel_file=openpyxl.Workbook() #  
    excel_sheet=excel_file.active #   
    excel_sheet.column_dimensions['A'].width = 100 #    & 
    excel_sheet.column_dimensions['B'].width = 20
    
    if sheetname != '':
        excel_sheet.title=sheetname #   default   title  
    for item in listdata:
        excel_sheet.append(item) # listdata     
    excel_file.save('tmp.xlsx') # 
    excel_file.close() # 
```

```python
import requests
from bs4 import BeautifulSoup

product_list=list()

for page_num in range(10):
    if page_num==0:
        res=requests.get('https://davelee-fun.github.io/')
    else:
        res=requests.get('https://davelee-fun.github.io/',str(page_num+1))
    soup=BeautifulSoup(res.content, 'html.parser')
    
    data=soup.select('div.card')
    for item in data:
        product_name=item.select_one('div.card-body h4.card-text')
        product_date=item.select_one('div.wrapfooter span.author-meta span.post-date')
        product_info=[product_name.get_text().strip(),product_date.get_text()]
        product_list.append(product_info)
write_excel_tmp('tmp.xlsx','',product_list)
```

```python
import requests
client_id='X_JRTQsqCYhl9NncUKCg'
client_pw='OSqB4Sfu_F'

start_num=1
num=0

excel_file=openpyxl.Workbook() #  
excel_sheet=excel_file.active #   
excel_sheet.column_dimensions['A'].width = 3 #    & 
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['rank','title','link'])
for index in range(10): # display  100 for    
    start_number=start_num+(index*100) # 1~100 101~200 ...
    naver_open_api='https://openapi.naver.com/v1/search/shop.json?query=&display=100&start='+str(start_number) # 1 , 101 , 201 , ... 
    header_params={'X-Naver-Client-Id':client_id,'X-Naver-Client-Secret':client_pw}
    res=requests.get(naver_open_api, headers=header_params)
    
    if res.status_code==200:
        data=res.json()
        #print(data['items'])
        for item in data['items']:
            num+=1 # num 1 100  for   , 101 200  for   , ...
            excel_sheet.append([num ,item['title'], item['link']])
        
    else :
        print("error:",res.status_code)
excel_file.save('shopping_with_.xlsx')
excel_file.close()
```
